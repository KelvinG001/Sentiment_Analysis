```python
import pandas as pd
import csv
from tqdm import tqdm

all_tweets = pd.read_excel('d:/卒論/data/TWEETS/tweets_info_combined.xlsx')
```

```python
import openai
import json

openai.api_key = "input your api key here"

def predict_GPT(tweet):
    """Predict sentiment of a tweet using OpenAI's Chat API."""
    try:
        # Ensure API key is set

        # Call OpenAI API
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",  
            messages=[
                {
                    "role": "user",
                    "content": f'Predict the sentiment of this tweet: "{tweet}". '
                               'Return a JSON object in the format: '
                               '{"sentiment": "positive/negative/neutral", "sentiment_score": 1(if positive)/0(if neutral)/-1(if negative)}. '
                               'Only return this JSON object and nothing else. Do not include anyother contents besides Json object or I cannot extract the Json successfully.',
                }
            ],
            max_tokens=150,
        )

        # Extract raw content
        content = response['choices'][0]['message']['content']
        # print(f"Raw API Response: {content}")

        # Ensure content is stripped of whitespace
        content = content.strip()
        

        # Parse JSON response
        try:
            json_content = json.loads(content)  # Parse as JSON
        except json.JSONDecodeError as e:
            print(f"JSON decoding failed: {e}. Using eval() as fallback.")
            json_content = eval(content)  # Fallback for dictionary-like strings

        return [json_content['sentiment'], json_content['sentiment_score']]
    
    except Exception as e:
        print(f"Error: {e}")
        return ["na", "na"]
```

```python
results = []

for _, tweet in tqdm(all_tweets[1625:80000].iterrows(), total = len(all_tweets[1625:80000]), desc = "Analysis Progress"):
    original = list(tweet)
    comment = tweet['content']
    re = predict_GPT(comment)

    original.extend(re)
    results.append(original)
```

```python
header = list(all_tweets.columns) + ["sentiment", "sentiment_score"]
with open(f'D:/卒論/data/sentiment_results_GPT.csv', 'w', newline='', encoding='utf-8-sig') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(header)  
    csv_writer.writerows(results)  

print("saved successfully!")
```
